```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('dataset/cleaned_combined_generation_dataset.csv')
dataset['timestamp'] = pd.to_datetime(dataset['timestamp'])
dataset.head()
```

```python
earlier_dataset=pd.read_csv('dataset/24-hours Delhi Power Consumption dataset.csv')
earlier_dataset['timestamp'] = pd.to_datetime(earlier_dataset['timestamp'])
earlier_dataset.head()
```

```python
# Merge the earlier_dataset into dataset on the 'timestamp' column
combined_dataset = pd.merge(dataset, earlier_dataset[['timestamp', 'solar_generation', 'load','compensation_method']], on='timestamp', how='left')

# Display the first few rows of the combined dataset
combined_dataset.head()
```

```python
combined_dataset_2023 = combined_dataset[combined_dataset['timestamp'].dt.year == 2023]
combined_dataset_2023.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

combined_dataset_2023.drop(
    columns=["timestamp", "electricity_demand", "population"], inplace=True
)

le = LabelEncoder()
combined_dataset_2023["compensation_method"] = le.fit_transform(
    combined_dataset_2023["compensation_method"]
)

# Split the dataset
train_data, test_data = train_test_split(
    combined_dataset_2023, test_size=0.2, random_state=42
)
train_data, val_data = train_test_split(train_data, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(
    train_data.drop(
        columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]
    )
)
val_data_scaled = scaler.transform(
    val_data.drop(
        columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]
    )
)
test_data_scaled = scaler.transform(
    test_data.drop(
        columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]
    )
)
```

```python
# Convert scaled arrays back to DataFrames
train_data_scaled_df = pd.DataFrame(train_data_scaled, columns=train_data.drop(columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]).columns)
val_data_scaled_df = pd.DataFrame(val_data_scaled, columns=val_data.drop(columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]).columns)
test_data_scaled_df = pd.DataFrame(test_data_scaled, columns=test_data.drop(columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]).columns)

# CNN Features
cnn_features = [
    "temperature",
    "wspd",
    "prcp",
    "rhum",
]
X_train_cnn = train_data_scaled_df[cnn_features].values
X_val_cnn = val_data_scaled_df[cnn_features].values
X_test_cnn = test_data_scaled_df[cnn_features].values

# LSTM Features
lstm_features = [
    "day_of_week",
    "hour_of_day",
    "season",
]
X_train_lstm = train_data_scaled_df[lstm_features].values
X_val_lstm = val_data_scaled_df[lstm_features].values
X_test_lstm = test_data_scaled_df[lstm_features].values

# ANN Features
ann_features = ["is_holiday", "is_weekend", "solar_generation"]
X_train_ann = train_data_scaled_df[ann_features].values
X_val_ann = val_data_scaled_df[ann_features].values
X_test_ann = test_data_scaled_df[ann_features].values

# Target Variable
y_train = train_data["load"]
y_val = val_data["load"]
y_test = test_data["load"]

# Reshape data for CNN and LSTM
X_train_cnn = np.array(X_train_cnn).reshape(
    (X_train_cnn.shape[0], X_train_cnn.shape[1], 1)
)
X_val_cnn = np.array(X_val_cnn).reshape((X_val_cnn.shape[0], X_val_cnn.shape[1], 1))
X_test_cnn = np.array(X_test_cnn).reshape((X_test_cnn.shape[0], X_test_cnn.shape[1], 1))

X_train_lstm = np.array(X_train_lstm).reshape(
    (X_train_lstm.shape[0], X_train_lstm.shape[1], 1)
)
X_val_lstm = np.array(X_val_lstm).reshape((X_val_lstm.shape[0], X_val_lstm.shape[1], 1))
X_test_lstm = np.array(X_test_lstm).reshape(
    (X_test_lstm.shape[0], X_test_lstm.shape[1], 1)
)
```

```python
# Convert scaled arrays back to DataFrames
train_data_scaled_df_cls = pd.DataFrame(train_data_scaled, columns=train_data.drop(columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]).columns)
val_data_scaled_df_cls = pd.DataFrame(val_data_scaled, columns=val_data.drop(columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]).columns)
test_data_scaled_df_cls = pd.DataFrame(test_data_scaled, columns=test_data.drop(columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]).columns)

# Target Variable
y_train_cls = train_data["compensation_method"]
y_val_cls = val_data["compensation_method"]
y_test_cls = test_data["compensation_method"]

# CNN Features
cnn_features_cls = [
    "temperature",
    "wspd",
    "prcp",
    "rhum",
]
X_train_cnn_cls = train_data_scaled_df_cls[cnn_features_cls].values
X_val_cnn_cls = val_data_scaled_df_cls[cnn_features_cls].values
X_test_cnn_cls = test_data_scaled_df_cls[cnn_features_cls].values

# LSTM Features
lstm_features_cls = [
    "day_of_week",
    "hour_of_day",
    "season",
]
X_train_lstm_cls = train_data_scaled_df_cls[lstm_features_cls].values
X_val_lstm_cls = val_data_scaled_df_cls[lstm_features_cls].values
X_test_lstm_cls = test_data_scaled_df_cls[lstm_features_cls].values

# ANN Features
ann_features_cls = ["is_holiday", "is_weekend", "solar_generation"]
X_train_ann_cls = train_data_scaled_df_cls[ann_features_cls].values
X_val_ann_cls = val_data_scaled_df_cls[ann_features_cls].values
X_test_ann_cls = test_data_scaled_df_cls[ann_features_cls].values

# Reshape data for CNN and LSTM
X_train_cnn_cls = np.array(X_train_cnn_cls).reshape(
    (X_train_cnn_cls.shape[0], X_train_cnn_cls.shape[1], 1)
)
X_val_cnn_cls = np.array(X_val_cnn_cls).reshape(
    (X_val_cnn_cls.shape[0], X_val_cnn_cls.shape[1], 1)
)
X_test_cnn_cls = np.array(X_test_cnn_cls).reshape(
    (X_test_cnn_cls.shape[0], X_test_cnn_cls.shape[1], 1)
)

X_train_lstm_cls = np.array(X_train_lstm_cls).reshape(
    (X_train_lstm_cls.shape[0], X_train_lstm_cls.shape[1], 1)
)
X_val_lstm_cls = np.array(X_val_lstm_cls).reshape(
    (X_val_lstm_cls.shape[0], X_val_lstm_cls.shape[1], 1)
)
X_test_lstm_cls = np.array(X_test_lstm_cls).reshape(
    (X_test_lstm_cls.shape[0], X_test_lstm_cls.shape[1], 1)
)
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import (
    Dense,
    LSTM,
    Conv1D,
    MaxPooling1D,
    Flatten,
    Dropout,
    Input,
    concatenate,
)

# CNN Model
cnn_input = Input(shape=(len(cnn_features), 1))
cnn_layer = Conv1D(filters=128, kernel_size=3, activation="relu")(cnn_input)
cnn_layer = MaxPooling1D(pool_size=2)(cnn_layer)
cnn_layer = Flatten()(cnn_layer)
cnn_output = Dense(128, activation="relu")(cnn_layer)

# LSTM Model
lstm_input = Input(shape=(len(lstm_features), 1))
lstm_layer = LSTM(100, activation="relu", return_sequences=True)(lstm_input)
lstm_layer = LSTM(50, activation="relu")(lstm_layer)

# Combine CNN output with LSTM features
combined_input = concatenate([cnn_output, lstm_layer])

# ANN Model
ann_input = Input(shape=(len(ann_features),))
combined_input = concatenate([combined_input, ann_input])
ann_layer = Dense(100, activation="relu")(combined_input)
ann_layer = Dropout(0.5)(ann_layer)
ann_layer = Dense(50, activation="relu")(ann_layer)
ann_output = Dense(1, activation="linear")(ann_layer)

# Combined Model
model = Model(inputs=[cnn_input, lstm_input, ann_input], outputs=ann_output)
model.compile(optimizer="adam", loss="mean_squared_error", metrics=["mae"])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit([X_train_cnn, X_train_lstm, X_train_ann], y_train, validation_data=([X_val_cnn, X_val_lstm, X_val_ann], y_val), epochs=50, batch_size=32, callbacks=[early_stopping])
```

```python
#building cnn model
cnn_input_cls = Input(shape=(len(cnn_features_cls), 1))
cnn_layer_cls = Conv1D(filters=128, kernel_size=3, activation='relu')(cnn_input_cls)
cnn_layer_cls = MaxPooling1D(pool_size=2)(cnn_layer_cls)
cnn_layer_cls = Flatten()(cnn_layer_cls)
cnn_output_cls = Dense(128, activation='relu')(cnn_layer_cls)

# LSTM Model
lstm_input_cls = Input(shape=(len(lstm_features_cls), 1))
lstm_layer_cls = LSTM(100, activation='relu', return_sequences=True)(lstm_input_cls)
lstm_layer_cls = LSTM(50, activation='relu')(lstm_layer_cls)

# Combine CNN output with LSTM features
combined_input_cls = concatenate([cnn_output_cls, lstm_layer_cls])

# ANN Model
ann_input_cls = Input(shape=(len(ann_features_cls),))
combined_input_cls = concatenate([combined_input_cls, ann_input_cls])
ann_layer_cls = Dense(100, activation='relu')(combined_input_cls)
ann_layer_cls = Dropout(0.5)(ann_layer_cls)
ann_layer_cls = Dense(50, activation='relu')(ann_layer_cls)
ann_output_cls = Dense(5, activation='softmax')(ann_layer_cls)  # Final layer with softmax activation

# Combined Model
model_cls = Model(inputs=[cnn_input_cls, lstm_input_cls, ann_input_cls], outputs=ann_output_cls)
model_cls.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping_cls = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history_cls = model_cls.fit([X_train_cnn_cls, X_train_lstm_cls, X_train_ann_cls], y_train_cls, validation_data=([X_val_cnn_cls, X_val_lstm_cls, X_val_ann_cls], y_val_cls), epochs=50, batch_size=32, callbacks=[early_stopping_cls])
```

```python
# Evaluate the model
loss, accuracy = model_cls.evaluate([X_test_cnn_cls, X_test_lstm_cls, X_test_ann_cls], y_test_cls)
print(f'Model Accuracy: {accuracy:.2f}')

# Make predictions
y_pred_cls = model_cls.predict([X_test_cnn_cls, X_test_lstm_cls, X_test_ann_cls])
y_pred_classes_cls = np.argmax(y_pred_cls, axis=1)

# Print classification report
from sklearn.metrics import classification_report
print(classification_report(y_test_cls, y_pred_classes_cls, target_names=le.classes_))
```
