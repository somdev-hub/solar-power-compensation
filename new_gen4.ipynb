```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('dataset/cleaned_combined_generation_dataset.csv')
dataset['timestamp'] = pd.to_datetime(dataset['timestamp'])
dataset.head()
```

```python
earlier_dataset=pd.read_csv('dataset/24-hours Delhi Power Consumption dataset.csv')
earlier_dataset['timestamp'] = pd.to_datetime(earlier_dataset['timestamp'])
earlier_dataset.head()
```

```python
# Merge the earlier_dataset into dataset on the 'timestamp' column
combined_dataset = pd.merge(dataset, earlier_dataset[['timestamp', 'solar_generation', 'load','compensation_method']], on='timestamp', how='left')

# Display the first few rows of the combined dataset
combined_dataset.head()
```

```python
combined_dataset_2023 = combined_dataset[combined_dataset['timestamp'].dt.year == 2023]
combined_dataset_2023.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

combined_dataset_2023.drop(
    columns=["timestamp", "electricity_demand", "population"], inplace=True
)

le = LabelEncoder()
combined_dataset_2023["compensation_method"] = le.fit_transform(
    combined_dataset_2023["compensation_method"]
)

# Split the dataset
train_data, test_data = train_test_split(
    combined_dataset_2023, test_size=0.2, random_state=42
)
train_data, val_data = train_test_split(train_data, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(
    train_data.drop(
        columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]
    )
)
val_data_scaled = scaler.transform(
    val_data.drop(
        columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]
    )
)
test_data_scaled = scaler.transform(
    test_data.drop(
        columns=["compensation_method", "load", "dwpt", "pres", "wdir", "month"]
    )
)
```

```python
import pickle

with open("models/scaler_ann.pkl", "wb") as f:
    pickle.dump(scaler, f)
    
with open("models/label_encoder_ann.pkl", "wb") as f:
    pickle.dump(le, f)
```

```python
X_train = train_data[
    [
        "temperature",
        "wspd",
        "prcp",
        "rhum",
        "day_of_week",
        "hour_of_day",
        "season",
        "is_holiday",
        "is_weekend",
        "solar_generation",
    ]
].values
X_test = test_data[
    [
        "temperature",
        "wspd",
        "prcp",
        "rhum",
        "day_of_week",
        "hour_of_day",
        "season",
        "is_holiday",
        "is_weekend",
        "solar_generation",
    ]
].values
X_val = val_data[
    [
        "temperature",
        "wspd",
        "prcp",
        "rhum",
        "day_of_week",
        "hour_of_day",
        "season",
        "is_holiday",
        "is_weekend",
        "solar_generation",
    ]
].values

y_train = train_data["load"].values
y_test = test_data["load"].values
y_val = val_data["load"].values
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


reg_model = Sequential()
reg_model.add(Dense(128, activation="relu", input_shape=(X_train.shape[1],)))
reg_model.add(Dense(64, activation="relu"))
reg_model.add(Dense(32, activation="relu"))
reg_model.add(Dropout(0.2))

reg_model.add(Dense(1, activation="linear"))

reg_model.compile(optimizer="adam", loss="mean_squared_error", metrics=["mae"])
```

```python
X_train_cls = train_data[
    [
        "temperature",
        "wspd",
        "prcp",
        "rhum",
        "day_of_week",
        "hour_of_day",
        "season",
        "is_holiday",
        "is_weekend",
        "solar_generation",
    ]
].values
X_test_cls = test_data[
    [
        "temperature",
        "wspd",
        "prcp",
        "rhum",
        "day_of_week",
        "hour_of_day",
        "season",
        "is_holiday",
        "is_weekend",
        "solar_generation",
    ]
].values
X_val_cls = val_data[
    [
        "temperature",
        "wspd",
        "prcp",
        "rhum",
        "day_of_week",
        "hour_of_day",
        "season",
        "is_holiday",
        "is_weekend",
        "solar_generation",
    ]
].values

y_train_cls = train_data["compensation_method"].values
y_test_cls = test_data["compensation_method"].values
y_val_cls = val_data["compensation_method"].values
```

```python
model_cls = Sequential()
model_cls.add(Dense(128, activation="relu", input_shape=(X_train_cls.shape[1],)))
model_cls.add(Dense(64, activation="relu"))
model_cls.add(Dense(32, activation="relu"))
model_cls.add(Dropout(0.2))
model_cls.add(Dense(5, activation="softmax"))

model_cls.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
reg_model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, batch_size=32, verbose=1)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor="val_loss", patience=10,restore_best_weights=True)

model_cls.fit(X_train_cls, y_train_cls, validation_data=(X_val_cls, y_val_cls), epochs=100, batch_size=32, verbose=1,callbacks=[early_stopping])
```

```python
# Evaluate the model
loss, accuracy = model_cls.evaluate(X_test_cls, y_test_cls)
print(f'Model Accuracy: {accuracy:.2f}')

# Make predictions
y_pred_cls = model_cls.predict(X_test_cls)
y_pred_classes_cls = np.argmax(y_pred_cls, axis=1)

# Print classification report
from sklearn.metrics import classification_report
print(classification_report(y_test_cls, y_pred_classes_cls, target_names=le.classes_))
```

```python
# Save the model
reg_model.save('models/combined_model_ann.h5')
model_cls.save('models/combined_model_cls_ann.h5')
```
