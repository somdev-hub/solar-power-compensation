```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
demand_dataset=pd.read_csv('dataset/hourly data(2000-2023).csv')
demand_dataset['timestamp']=pd.to_datetime(demand_dataset['timestamp'])
# demand_dataset.set_index('timestamp',inplace=True)
demand_dataset.head()
```

```python
weather_cleaned_dataset=pd.read_csv('dataset/weather_data_delhi_2000_2023_cleaned.csv')
weather_cleaned_dataset['timestamp'] = pd.to_datetime(weather_cleaned_dataset['timestamp'])

demand_dataset = pd.merge(demand_dataset, weather_cleaned_dataset[['timestamp', 'dwpt', 'wspd', 'wdir', 'prcp', 'rhum', 'pres']], on='timestamp', how='left')

demand_dataset.head(20)
# combined_dataset = pd.merge(demand_dataset, weather_cleaned_dataset, on='timestamp', how='inner')

# combined_dataset.head()
```

```python
demand_dataset.isna().sum()
```

```python
# Fill NaN values for specific columns with the average of the previous 3 hours
demand_dataset = demand_dataset.set_index('timestamp')
columns_to_fill = ['dwpt', 'wspd', 'wdir', 'prcp', 'rhum', 'pres']
demand_dataset[columns_to_fill] = demand_dataset[columns_to_fill].rolling(window='3H').mean().fillna(method='bfill')
demand_dataset = demand_dataset.reset_index()

# Display the first few rows to verify
demand_dataset.head(20)
```

```python
demand_dataset.isna().sum()
```

```python
demand_dataset.drop(columns=['solar_generation'], inplace=True)
```

```python
demand_dataset.head()
```

```python
demand_dataset.columns
```

```python
'''
The electricity demand forecasting will employ three models i.e. CNN, LSTM and ANN
Group I data for CNN => ['temperature','dwpt','wspd','wdir','prcp','rhum','pres']
Group II data fro LSTM => ['day_of_week','hour_of_day','season','month','output from CNN']
Group III data for ANN => ['is_holiday','is_weekend','output from LSTM']

combining the data of the 3 models the forecasting of the next 24 hours of electricity demand will be done
'''
```

```python
demand_dataset['season'] = demand_dataset['timestamp'].dt.month % 12 // 3 + 1

demand_dataset['month'] = demand_dataset['timestamp'].dt.month

demand_dataset.head()
```

```python
hourly_population=pd.read_csv('dataset/hourly_population.csv')
hourly_population['year']=pd.to_datetime(hourly_population['year'])

hourly_population.rename(columns={'year':'timestamp'},inplace=True)

hourly_population.head()
```

```python
hourly_population.isna().sum()
```

```python
demand_dataset = pd.merge(demand_dataset, hourly_population, on='timestamp', how='left')

demand_dataset.head()
```
